```python
## Loading the required modules

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the data

df = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')
df.shape
```

```python
## Scatter Plot

sns.relplot(data = df, x = 'Height', y = 'Weight')
```

```python
## Lineplot

x = np.random.rand(5)
y = 5 + np.random.rand(5)*x

x_y = pd.DataFrame({'x':x, 'y':y})

sns.lineplot(data = x_y, x = 'x', y = 'y')
```

```python
## Bar Charts

index_counts = df.groupby(['Index']).size().reset_index()
index_counts.columns = ['Index', 'Counts']

sns.barplot(data = index_counts, x = 'Index', y = 'Counts')
```

```python
## Stacked Bar Charts

index_gender_counts = df.groupby(['Index','Gender']).size().reset_index()
index_gender_counts.columns = ['Index', 'Gender', 'Counts']

index_gender_counts.set_index('Index').pivot(columns = 'Gender').plot(kind = 'bar', stacked = True)
```

```python
## Column bar charts

sns.barplot(data = index_gender_counts, x = 'Index', y = 'Counts', hue = 'Gender')
```

```python
## Histograms

sns.histplot(data = df, x = 'Height')
```

```python
## KDE Plot

sns.kdeplot(data = df, x = 'Height')
```

```python
## Box Plots

sns.boxplot(data = df, x = 'Index', y = 'Height')
```

### Multiple Plots

```python
## scatter plots : height vs weight by Gender

sns.relplot(data = df, x = 'Height', y = 'Weight', col = 'Gender', hue = 'Gender')
```

```python
sns.relplot(data = df, x = 'Height', y = 'Weight', hue = 'Gender')
```

```python
## Visualising data by two categories

grid = sns.FacetGrid(data = df, col = 'Index', row = 'Gender', margin_titles = True)

grid.map(plt.hist, 'Height')
```
